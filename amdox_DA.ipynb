DATA CLEANING OF BITCOIN


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/drive/MyDrive/bitcoin_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace=True)
```

```python
df.info()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Close'])
plt.title("Boxplot of Bitcoin Closing Price")
plt.show()
```

```python
Q1 = df['Close'].quantile(0.25)
Q3 = df['Close'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Close'] < lower) | (df['Close'] > upper)]
outliers
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Close'])
plt.title("Bitcoin Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Volume'], bins=30, kde=True)
plt.title("Bitcoin Volume Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```

```python
df['Daily_Return'] = df['Close'].pct_change()
df.dropna(inplace=True)
plt.figure(figsize=(12,6))
plt.plot(df['Daily_Return'])
plt.title("Bitcoin Daily Returns")
plt.show()
```

```python
df.to_csv("BTC_cleaned.csv")
```

Data Analysis


```python
df["Close"].plot(figsize=(10,5))
plt.title("Closing Price Trend")
plt.show()
```

```python
df["MA_7"] = df["Close"].rolling(window=7).mean()
df["MA_30"] = df["Close"].rolling(window=30).mean()
df[["Close", "MA_7", "MA_30"]].plot(figsize=(10,5))
```

```python
df["Volatility"] = df["Daily_Return"].rolling(7).std()
df["Volatility"].plot(figsize=(10,5))
```

ML Models


```python
train = df.iloc[:int(len(df)*0.8)]
test = df.iloc[int(len(df)*0.8):]
```

ARIMA

```python
!pip install pmdarima
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train["Close"], order=(2,1,2))
model_fit = model.fit()

forecast = model_fit.forecast(steps=len(test))
```

```python
plt.plot(test.index, test["Close"], label="Actual")
plt.plot(test.index, forecast, label="Predicted")
plt.legend()
plt.show()
```

Prophet

```python
from prophet import Prophet

prophet_df = df.reset_index()[["Date","Close"]]
prophet_df.columns = ["ds","y"]

model = Prophet()
model.fit(prophet_df)

future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

model.plot(forecast)
```
